$\pi_{i} P_{ij}=\pi_{j} P_{ij}$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x,mu,std):
  return (1/np.sqrt(2*np.pi*std**2))*np.exp(-(x-mu)**2/(2*std**2))
```

```python
mu= 0.2
sigma=0.1
```

```python
delta=0.5
```

```python
n=50000
x=np.zeros(n)
x[0]=0
acept=0
for i in range(0,n-1):
  y=x[i]+np.random.uniform(-delta,delta)
  if np.random.rand()<min(1,f(y,mu,sigma)/f(x[i],mu,sigma)):
    x[i+1]=y
    acept+=1
  else:
    x[i+1]=x[i]
print("la aceptacion fue del ",acept/n*100,"%")
```

```python
plt.hist(x,density=True, bins=30)
xs=np.linspace(-1,1,100)
plt.plot(xs,f(xs,mu,sigma))
```

```python
print(np.mean(x))
print(np.std(x))
```

```python
plt.plot(x[:100])#los primeros 100 valores aleatorios
```

#Para poisson


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, lamb):
  return (lamb**x * np.exp(-lamb)) / np.math.factorial(int(x))
```

```python
lamb = 5
delta = 10
```

```python
n = 50000
x = np.zeros(n)
x[0] = 0
acept = 0
for i in range(0, n-1):
  y = x[i] + np.random.uniform(-delta, delta)
  if y >= 0:
    if np.random.rand() < min(1, f(y, lamb) / f(x[i], lamb)):
      x[i+1] = y
      acept += 1
    else:
      x[i+1] = x[i]
  else:
    x[i+1] = x[i]

print("Tasa de aceptacion:", acept/n*100, "%")
```

```python
plt.hist(x, density=True, bins=30)
xs = np.arange(0, 20)
plt.plot(xs, f(xs,lamb))
```

```python
print(np.mean(x))
print(np.std(x))
```
