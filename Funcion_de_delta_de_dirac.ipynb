```python
import sympy as sp
sp.init_printing()
```

```python
S=sp.Symbol("S")
c=sp.Symbol("c")
t=sp.Symbol("t")
```

```python
(t+4)**2
```

```python
((t+4)**2).expand()
```

```python
((t+4)**2).subs(t,1)
```

```python
((t**+2)*(t**2+8)).expand()
```

```python
sp.factor(t**2+6*t+5)
```

```python
sp.simplify(t**2+2*t**2+4*t+5)
```

```python
sp.plotting.plot(t*t)
```

```python
sp.plotting.plot(10*sp.exp(0.5*t))
```

```python
S=sp.Function("S")
ode=sp.Eq(sp.Derivative(S(t),t)-c*S(t),0)
ode
sol=sp.dsolve(ode,S(t),ics={S(0):10000})
sol.subs({c:0.07})
```

```python
Y=sp.Heaviside(t-sp.pi)
```

```python
u=t**2+(t-t**2)*Y
```

```python
sp.plotting.plot(u)
```

```python
U=1/2*sp.Heaviside(t-2)-1/2*sp.Heaviside(t-4)
sp.plotting.plot(U)
```

```python
D=sp.DiracDelta(t-3)
sp.plotting.plot(D)
```
