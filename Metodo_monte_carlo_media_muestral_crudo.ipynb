```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parámetros
a, b = 0, 1
n_simulaciones = 100
tamano_muestra = 20
```

```python
def funcion_integral(x):
    #Función que queremos evaluar para estimar la media muestral
    return np.cos(np.pi * x / 2)

def monte_carlo_media_muestral(funcion, a, b, n_simulaciones, tamano_muestra):

    resultados = []
    for _ in range(n_simulaciones):
        # Generar una muestra aleatoria en el intervalo [a, b]
        muestra_x = np.random.uniform(a, b, tamano_muestra)
        # Evaluar la función en cada punto de la muestra
        muestra_y = funcion(muestra_x)
        # Calcular la media de la muestra
        media_muestral = np.mean(muestra_y)
        resultados.append(media_muestral)

    # Estimar la media muestral
    media_estimacion = np.mean(resultados)
    return media_estimacion, resultados
```

```python
# Estimación de la media muestral
media_estimacion, resultados = monte_carlo_media_muestral(funcion_integral, a, b, n_simulaciones, tamano_muestra)
print(f"Estimación de la media muestral: {media_estimacion}")
```

```python
# Crear una serie de puntos para graficar la función objetivo
x = np.linspace(a, b, 500)
y = funcion_integral(x)

# Graficar los resultados
plt.figure(figsize=(10, 6))

# Histograma de las medias muestrales
plt.hist(resultados, bins=30, edgecolor='k', alpha=0.7, label='Medias Muestrales')

# Graficar la función objetivo en el mismo gráfico
plt.plot(x, y, label='Función Objetivo: cos(πx/2)', color='b')

plt.title('Distribución de las Medias Muestrales y Función Objetivo')
plt.xlabel('Valor')
plt.ylabel('Frecuencia / Valor de la Función')
plt.legend()
plt.grid(True)
plt.show()
```
