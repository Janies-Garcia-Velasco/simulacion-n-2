```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Número de muestras
n_muestras = 1000
```

```python
# Definir la función que queremos integrar
def funcion(x):
    return np.cos(np.pi * x / 2)
```

```python
# Muestreo de importancia: generamos muestras según una distribución
# En este caso, usaremos una distribución uniforme en el intervalo [0, 1]
muestras_x = np.random.uniform(0, 1, n_muestras)

# Calcular el valor de la función en cada muestra
valores_funcion = funcion(muestras_x)

# Estimar la integral usando el promedio de las evaluaciones
estimacion_integral = np.mean(valores_funcion)
```

```python
# Varianza y error estimado
varianza_estimado = np.var(valores_funcion) / n_muestras
error_estimado = np.sqrt(varianza_estimado)

print(f'Estimación de la integral: {estimacion_integral:.4f}')
print(f'Error estimado: {error_estimado:.4f}')
```

```python
# Graficar los resultados
x = np.linspace(0, 1, 400)
y = funcion(x)

plt.figure(figsize=(12, 6))

# Graficar la función objetivo
plt.plot(x, y, label='Función objetivo: cos(pi*x/2)', color='blue')

# Graficar el histograma de los valores de la función
plt.hist(valores_funcion, bins=30, density=True, alpha=0.6, color='green', edgecolor='black', label='Distribución de los valores')

plt.title('Distribución de los valores de la función junto con la función objetivo')
plt.xlabel('Valor de la función')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(True)
plt.show()
```
