```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parámetros para el método de aceptación y rechazo
muestras = 1000  # Número de muestras aleatorias
a, b = 0, 1  # Intervalo de integración
y = 1  # Valor máximo de la función f(x) (es la cota superior)
```

```python
# Definir la función que queremos integrar
def f(x):
    return np.cos(np.pi * x/2)
```

```python
# Generar puntos aleatorios en el intervalo [0, 1] para x
x_random = np.random.uniform(a, b, muestras)

# Generar puntos aleatorios en el intervalo [0, y_max] para y
y_random = np.random.uniform(0, b, muestras)

# Evaluar la función f en los puntos x_random
f_x = f(x_random)

# Aplicar el criterio de aceptación: solo aceptar puntos donde y_random < f(x_random)
puntos_aceptados = y_random < f_x

# Calcular la estimación de la integral
# Proporción de puntos aceptados multiplicada por el área del rectángulo de muestreo
area_estimada = np.sum(puntos_aceptados) / muestras * (b - a) * y

# Graficar los resultados
plt.figure(figsize=(10, 6))
x_plot = np.linspace(a, b, 1000)
plt.plot(x_plot, f(x_plot), 'r-')
plt.scatter(x_random[puntos_aceptados], y_random[puntos_aceptados], color='green', label='Puntos Aceptados')
plt.scatter(x_random[~puntos_aceptados], y_random[~puntos_aceptados], color='blue', label='Puntos Rechazados')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Método de Aceptación y Rechazo Monte Carlo')
plt.legend()
plt.grid(True)
plt.show()

print(f'Estimación de la integral: {area_estimada}')
```
