```python
import random
import math

# Función para generar un tiempo aleatorio exponencial
def generar_tiempo_exponencial(lambd):
    return -math.log(random.uniform(0, 1)) / lambd

def simular(lambd_llegadas, lambd_servicio_1, lambd_servicio_2, T0):
    # Inicializar variables
    t = 0
    c1 = 0  # Contador para clientes que terminan servicio 1
    c2 = 0  # Contador para clientes que terminan servicio 2
    Nll = 0  # Contador de llegadas
    SS = []  # Estado del sistema (lista de clientes en servicio)

    # Listas para registrar tiempos de llegada y salida
    llegadas = []
    salidas = {}

    # Generar el primer tiempo de llegada (T0)
    tll = T0
    t1 = float('inf')  # Inicializamos t1 como infinito
    t2 = float('inf')  # Inicializamos t2 como infinito

    while True:
        # Caso 1: Llegada de un cliente
        if tll <= t1 and tll <= t2:
            print("Caso 1: Llegada de un cliente.")
            t = tll
            Nll += 1
            llegadas.append(t)  # Registrar la llegada del cliente

            # Generar nuevo tiempo de llegada
            Tt = generar_tiempo_exponencial(lambd_llegadas)
            tll = t + Tt  # Próxima llegada

            # Actualización de SS dependiendo del estado
            if len(SS) == 0:  # No hay nadie en el sistema
                SS = [1, Nll, 0]  # Nuevo cliente entra al servicio 1
                Y1 = generar_tiempo_exponencial(lambd_servicio_1)
                t1 = t + Y1
            elif SS[0] == 1 and len(SS)>2 and SS[2]==0:  # Un cliente en servicio 1, nadie en servicio 2
                SS = [2, SS[1], Nll]  # Cliente pasa al servicio 2
                Y2 = generar_tiempo_exponencial(lambd_servicio_2)
                t2 = t + Y2
            elif SS[0] == 1 and SS[2] != 0:  # Un cliente en servicio 1 y en servicio 2
                SS[0] = 2
                SS.append(Nll)  # Nuevo cliente al servicio 1
                Y1 = generar_tiempo_exponencial(lambd_servicio_1)
                t1 = t + Y1
            elif SS[0] > 1:  # Más de un cliente en el sistema
                SS[0] += 1
                SS.append(Nll)  # Agregar cliente al sistema

        # Caso 2: Cliente termina servicio 1 y pasa al servicio 2
        elif t1 <= tll and t1 <= t2:
            print("Caso 2: Cliente pasa del servicio 1 al servicio 2.")
            t = t1
            c1 += 1

            if SS:
              if len(SS)>=2:
                salidas[SS[1]]=t
            else:
              print("Cuidado, SS es mas pequeña de dos elementos. La salida es muy pequeña")

            if SS[0] == 1:  # Solo un cliente, sistema vacío después del servicio 1
                SS = []
                t1 = float('inf')
            elif SS[0] == 2:  # Un cliente en servicio 2
                SS = [1, 0, SS[2]]
                t1 = float('inf')
            else:  # Más de un cliente en el sistema
                SS = [SS[0] - 1] + SS[2:]  # Eliminar el cliente que salió
                Y1 = generar_tiempo_exponencial(lambd_servicio_1)
                t1 = t + Y1

        # Caso 3: Cliente termina servicio 2 y sale del sistema
        elif t2 <= tll and t2 <= t1:
            print("Caso 3: Cliente termina el servicio 2 y sale del sistema.")
            t = t2
            c2 += 1

            if len(SS)>2:  # Solo un cliente en el sistema
                salidas[SS[2]]=t
            else:
              print("SS es muy corta")

            if SS[0]==1:
              SS=[]
              t2=float("inf")
            else:
              t2=float("inf")

        # Condición de salida (se puede modificar)
        if c2 >= 10:  # Detener después de 10 clientes que terminan servicio 2
            break

    # Retornar los registros de llegadas y salidas
    return llegadas, salidas

# Parámetros de ejemplo
lambd_llegadas = 1.0  # Tasa de llegada
lambd_servicio_1 = 1.5  # Tasa de servicio 1
lambd_servicio_2 = 2.0  # Tasa de servicio 2
T0 = generar_tiempo_exponencial(lambd_llegadas)

# Ejecutar la simulación
llegadas, salidas = simular(lambd_llegadas, lambd_servicio_1, lambd_servicio_2, T0)

# Imprimir los resultados
print(f"Llegadas: {llegadas}")
print(f"Salidas: {salidas}")
```
